# Splitting data for regression

```python
from sklearn.model_selection import train_test_split

#Splitting the dataset into training & testing data
x, y = df[['Month', 'Max Temp 0 C', 'Min Temp 0 C',
       'Relative Humidity %', 'Wind Speed (m/s)','Cloud Coverage %', 'Bright Sunshine (Hours)', 'Station Number']], df[['Rainfall (mm)']]
       
#train=80%, test= 20%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#saving the split data index
train = x_train.index
test = x_test.index
training_index = pd.DataFrame(train)
testing_index = pd.DataFrame(test)
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
# define a fuction
lr=LinearRegression()
# fit the model with split dataset
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr_prediction_test = lr.predict(x_test)
lr_prediction_train = lr.predict(x_train)
```

```python
#root mean square error to check accuracy
lr_rmse_train = np.sqrt(mean_squared_error(y_train, lr_prediction_train))
print("train=",lr_rmse_train)
lr_rmse_test = np.sqrt(mean_squared_error(y_test, lr_prediction_test))
print("test=",lr_rmse_test)
```

```python
##cross validation
from sklearn.model_selection import cross_val_score
mse=cross_val_score(lr,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
np.mean(mse)
```

```python
#r2 score
lr_r2_train = r2_score(y_train, lr_prediction_train)
print("train=",lr_r2_train)
lr_r2_test = r2_score(y_test, lr_prediction_test)
print("test=",lr_r2_test)
```

```python
##prediction 
lr_pred=lr.predict(x_test)
print(lr_pred)
```

```python
sns.displot(lr_pred-x_test,kind='kde');
```

##k-nearest neighbors(KNN)

```python
from sklearn.neighbors import KNeighborsRegressor

#Find the best value of hyper parameter
i = 1
k = 1
r = 0
while i<=100:
    KNN = KNeighborsRegressor(i,weights='distance',algorithm='auto')
    KNN.fit(x_train,y_train)
    y_prediction_test = KNN.predict(x_test)
    r2 = r2_score(y_test, y_prediction_test)
    if r2> r:
        k=i
    i += i
print("Best Value of k is: " ,k)
```

# Clustering-Silhouetter Coefficient

```python
# Imports
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
from sklearn.model_selection import train_test_split

#Splitting the dataset into training & testing data
x, y = df[['Month', 'Max Temp 0 C', 'Min Temp 0 C',
       'Relative Humidity %', 'Wind Speed (m/s)','Cloud Coverage %', 'Bright Sunshine (Hours)', 'Station Number']], df[['Rainfall (mm)']]
       
#train=80%, test= 20%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
def elbow_method(x, max_k=64):
    costs = []
    for k in range(2, max_k):
        model = KMeans(n_clusters=k)
        model.fit(x)
        costs.append(model.inertia_)
    plt.close()
    plt.plot(list(range(2, max_k)), costs)
    plt.xlabel("# of clusters (K)")
    plt.ylabel("Cost")
```

```python
elbow_method(x);
```
